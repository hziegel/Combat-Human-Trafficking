```python
import os
import random
import shutil

def create_data_subset(source_dir, dest_dir, subset_size):

    if not os.path.exists(dest_dir):
        os.makedirs(dest_dir)

    all_files = os.listdir(source_dir)
    num_files = len(all_files)

    if subset_size > num_files:
        return
        #raise ValueError("Subset size cannot be larger than the total number of files.")

    selected_files = random.sample(all_files, subset_size)

    for filename in selected_files:
        source_path = os.path.join(source_dir, filename)
        dest_path = os.path.join(dest_dir, filename)
        shutil.copy2(source_path, dest_path)  # Use copy2 to preserve metadata

if __name__ == "__main__":
    source_directory = "/content/drive/MyDrive/Combat-Human-Trafficking/train_images/"
    destination_directory = "/content/drive/MyDrive/Combat-Human-Trafficking/train_images_subset/"
    subset_size = 5  # Specify the desired number of files in the subset

    all_folders = os.listdir(source_directory)
    for fldr in all_folders:
      create_data_subset(source_directory + fldr, destination_directory + fldr, subset_size)
    print(f"Subset of {subset_size} files created in {destination_directory}")
```
